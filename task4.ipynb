```python
import pandas as pd
```

```python
x=pd.read_csv(r"D:\twitter_training.csv")
```

```python
x
```

```python
x.columns
```

```python
x.isnull().sum()
```

```python
x=x.dropna(subset=["TEXT"])
```

```python
x.loc[:,"TEXT"]=x["TEXT"].str.lower()
```

```python
x.loc[:,"TEXT"]=x["TEXT"].str.replace(r"http\\S+|www\\S+","",regex=True)
```

```python
x.loc[:,"TEXT"]=x["TEXT"].str.replace(r"[^\w\s]","",regex=True)
```

```python
x.loc[:,"TEXT"]=x["TEXT"].str.replace(r"\s+"," ",regex=True).str.strip()
```

```python
x.loc[:,"SENTIMENT"]=x["SENTIMENT"].str.strip().str.capitalize()
```

```python
x.loc[:,"TOPIC"]=x["TOPIC"].str.strip().str.lower()
```

```python
x=x.drop_duplicates()
```
